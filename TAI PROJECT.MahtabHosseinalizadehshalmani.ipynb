```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
%matplotlib inline
```

```python
!ls
```

```python
data = pd.read_excel('Reduced Features for TAI project.xlsx') #reading the excel file 
```

```python
data_ = data.copy() # keeping a copy of ecel file for other uses
```

```python
list_of_columns = list(data.columns) 
```

```python
data.head()
```

```python
data = data.drop('Patient ID', axis=1) # eliminating the Patient ID as we need to normalize our dataset and having patient id makes it nonsense
data.head()
```

```python
data.shape
```

```python
Y = data['Label']
X = data.drop('Label', axis=1) # defining the input (x) or our column in dataset  and output(y) or the label
```

```python
data['Label'].value_counts()
```

```python
X.head()
```

```python
min_max_scaler = preprocessing.MinMaxScaler() #normalizing the data to get them all the features between 0 and 1 to work with them 
X_scaled = min_max_scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=list_of_columns[1:-1])
```

```python
X_scaled
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size = 0.20) # traing the model and defining the size of training and testing sets
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
svm_classifier = SVC(kernel='linear') # choosing SVM as the ML model for this dataset 
svm_classifier.fit(X_train, Y_train)
```

```python
Y_prediction = svm_classifier.predict(X_test)
```

```python
print(confusion_matrix(Y_test, Y_prediction)) # getting data classified by SVM with 74% accuracy!!
print(classification_report(Y_test, Y_prediction))
```
